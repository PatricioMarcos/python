```python
# Markdown Header
# This is a Jupyter Notebook created with Visual Studio Code

import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
random_numbers = np.random.rand(100)

# Calculate quartiles
quartiles = np.percentile(random_numbers, [25, 50, 75])

# Plot histogram
plt.hist(random_numbers, bins=10)
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Plot boxplot
plt.boxplot(random_numbers)
plt.title('Boxplot of Random Numbers')
plt.ylabel('Value')
plt.show()
```
